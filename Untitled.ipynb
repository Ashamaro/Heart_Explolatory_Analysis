```python
#Handling outliers in boxplots of numerical data
#from columns chol(cholestrol)
def remove_outliers(df,column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5*IQR
  upper_bound = Q3 + 1.5*IQR
  return df[(df[column]>=lower_bound)&(df[column]<=upper_bound)]
data = remove_outliers(data,'chol')
```

```python
sns.boxplot(x = 'target',y = 'chol',data = data)
plt.title('DIstribution of chol by target')
plt.show()
```

```python
#Scaling of Numerical Featuers
#Z- Menthod will be used because outliers are present
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_features = ['age','chol','thalach','oldpeak','trestbps']
data = data.copy()
data[numerical_features] = scaler.fit_transform(data[numerical_features])
print(data.head())
```

```python
#Random Forest Feature Importance for  Numerical Features to Re_affirm
#manual analysis about these features
from sklearn.ensemble import RandomForestClassifier
numerical_features = ['age','chol','thalach','oldpeak','trestbps']
x = data[numerical_features]
y = data['target']
rf = RandomForestClassifier(random_state = 42)
rf.fit(x,y)
feature_importances = pd.Series(rf.feature_importances_,index = numerical_features)
feature_importance = feature_importances.sort_values(ascending = False,inplace = True)
print('Features Importance')
print(feature_importances)
```

```python
#Analyzing Categorical Data
# First step Completed (Countplots of each feature of Categorical Data)
#insights through countplots are that fbs,restecg have weak relation with target while sex,thal,slope,ca and cp have strong relation
#Second Step to affirm further which featutre is important is Cross Tab Analysis
categorical_features = ['sex','cp','fbs','restecg','exang','thal','ca']
for col in categorical_features:
  result = pd.crosstab(data[col],data['target'],normalize = 'index')*100
  print(f"Feature:{col}")
  print(result)
```

```python
#chi_square test (statiscal test to check which categorical feature has strong relation with target)
from scipy.stats import chi2_contingency
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
for col in categorical_features:
  contingency_table = pd.crosstab(data[col],data['target'])
  chi2,p,dof,expected = chi2_contingency(contingency_table)
  print(f"feature:{col}")
  print(f"Chi_Squred Statistics:{chi2}, P_Value:{p}")
  if p < 0.05:
    print(f"Feature {col} is significantly related to target")
  else:
    print(f"Feature {col} is not significantly related to target")
```

```python
data = data.drop('fbs',axis = 1)
```

```python
#Encoding of Categorical Features for Machine Learning
data = pd.get_dummies(data,columns = ['cp','sex','slope','ca','thal','exang'],drop_first = True)
```

```python
print(data.head())
```

```python
#Splitting the Dataset into test and train data
from sklearn.model_selection import train_test_split
x = data.drop(columns = ['target'])
y = data['target']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 101)
```

```python
#Training for Machine Learning Model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
percentage = accuracy*100
print(f"Acuuracy:,{percentage:.2f}%")
print("Classification Report:",classification_report(y_test,y_pred))
```
